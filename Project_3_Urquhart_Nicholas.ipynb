```python
def sortLog(entries):
    #The goal is to create a dictionary where the key is the date and its value is a sorted list of log info on that day
    dateDict = {}
    #30 days in may, so through each day, will iterate through log, looking for the same day, then store it into the dictionary
    for n in range(1, 31, 1):
        sortedFile = []
        #   ^  "sorted" as in the days are sorted, not time sorted
        #days that are single digit appear as doubledigit in the user log, so the string must have the 0 in front.
        if n < 10:
            num = "0" + str(n)
        else:
            num = str(n)
        #splitting the date into a list for iteration
        for i in range(0, len(entries) - 3, 4):
            dateList = entries[i].split(" ")
            dateSep = dateList[0].split("-")
            date = dateSep[2]
            if num in date:
                sortedFile.append(entries[i])
                sortedFile.append(entries[i + 1])
                sortedFile.append(entries[i + 2])
                sortedFile.append(entries[i + 3])
        dateDict.update({"05 - " + num : sortedFile})
    #now that we have a dictonary of the log, sorted by date, now sort by time
    
    #I have some prior programming experience, so I know bubble sort
    for sortedDates in dateDict.values():
        sorted = 0
        while sorted != 1:
            sorted = 1
            for i in range(4, len(sortedDates), 4):
                OdateList = sortedDates[i].split(" ")
                OdateSep = OdateList[1].split(":")
                OdateSep = intConvert(OdateSep)
                CdateList = sortedDates[i - 4].split(" ")
                CdateSep = CdateList[1].split(":")
                CdateSep = intConvert(CdateSep)
                found = 0
                c = 0
                while found != 1:
                    #means we didn't find any difference in date, so we do nothing 
                    if c == 3:
                        found = 1
                    #means we found that the first date is less than the the next, so we do nothing
                    elif OdateSep [c] > CdateSep[c]:
                        found = 1
                    #means we found that the first date is greater than the next, so we switch
                    elif OdateSep[c] < CdateSep[c]:
                        found = 1
                        sorted = 0
                        #didn't want to manipulate i, so we duplicate it
                        r = i
                        for r in range(r, r + 4, 1):
                            temp = sortedDates[r]
                            sortedDates[r] = sortedDates[r - 4]
                            sortedDates[r - 4] = temp
                    c += 1
    return dateDict

#this function is used for converting the sepeated dates list, which are strings, into ints 
def intConvert(sList):
    for i in range(0, len(sList), 1):
        sList[i] = int(sList[i])
    return sList
#this function is used for suspicous activities and it takes the log info of date and time and returns what hour it occurs
def detTime(date):
    timeList = date.split(" ")
    time = timeList[1].split(":")
    time = intConvert(time)
    if time[0] >= 0 and time[0] <= 5:
        return 1
    else:
        return 0
    
```
